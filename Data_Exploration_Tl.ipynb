```
                                                    #Data Exploration using Python
```

```
import pandas as pd
```

```
data=pd.read_csv('pokemon_data.csv')
```

```
#Printing first 5 rows

data.head(5)
```

```
#Printing the shape

data.shape
```

```
#Sorting name in alphabetical order

data.sort_values(['Name'],ascending=True)
```

```
#Checking for duplicate values

data.duplicated()
```

```
#Printing the names of pokemon

for index,row in data.iterrows():
    print(index,row['Name'])
```

```
#Prining names of pokemon whose defense is greater than 70

data.loc[data['Defense']>70]
```

```
#Sorting the values obtained above in Descending order of Defense

tempdata=data.loc[data['Defense']>70]

tempdata.sort_values(['Defense'],ascending=False)
```

```
#Printing the mean values of all pokemon

data.groupby(['Name']).mean()
```

```
#Creating a new column showing defense percentage

data['Defense Per']=data.iloc[:,4]*(1/100)

data.head(5)
```

```
#Importing matplotlib 
import matplotlib.pyplot as plt
```

```
graphdata=data.head(10)

x=graphdata['Defense']
```

```
#Plotting histogram

plt.hist(x,bins=10)
plt.title("Defense of 10 Pokemons")
plt.show()
```
